## Iterative policy evaluation in Small Gridworld


### 1. initalizing

```python
import matplotlib.pyplot as plt

SIZE = 4
INITIAL_VALUE = 0.0
REWARD = -1.0
ITERATION = 100
```

### 2. define gridworld class

```python
class gridworld:
  def __init__(self):
    self.size = SIZE
    self.reward = REWARD
    self.data = [[INITIAL_VALUE for _ in range(self.size)] for _ in range(self.size)]
    self.fig, self.ax = plt.subplots(1)
    self.ax.axis('tight')
    self.ax.axis('off')
    self.t = self.ax.table(cellText=self.data, loc='center')

  def update_table(self, new_data):
    self.data = new_data
    self.fig, self.ax = plt.subplots(1)
    self.ax.axis('tight')
    self.ax.axis('off')
    self.t = self.ax.table(cellText=self.data, loc='center')

  def eval(self):
    new_data = [[0.0 for _ in range(self.size)] for _ in range(self.size)]
    for i in range(self.size):
      for j in range(self.size):
        if (i==0 and j==0) or (i==self.size-1 and j==self.size-1):
          continue
        new_value = self.data[max(i-1,0)][j]
        new_value += self.data[min(i+1,self.size-1)][j]
        new_value += self.data[i][max(j-1,0)]
        new_value += self.data[i][min(j+1,self.size-1)]
        new_data[i][j] = round(self.reward + new_value / 4, 4)
    return new_data

  def show_table(self, n):
    self.t.scale(1,4)
    plt.title(f"k = {n}")
    plt.show()
```

### 3. iterative policy evaluation

```python
# init gridworld
model = gridworld()
model.show_table(0)

# evaluation and show results
for i in range(ITERATION):
  model.update_table(model.eval())
  model.show_table(i+1)
```
