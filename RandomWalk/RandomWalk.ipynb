#### Random Walk Example

#### 1. Initializng

```python
import random
import time

state_dict = {"A":0, "B":1, "C":2, "D":3, "E":4}
state = list(state_dict.keys())
value = [0.5, 0.5, 0.5, 0.5, 0.5]
alpha = 0.1
gamma = 1
```

#### 2. Functions

```python
# episode generator
def generate_episode():
  episode = ["C"]
  s_idx = 2
  while True:
    rand = (round(random.random(), 2)) * 100
    if rand % 2 == 0: # go right
      s_idx += 1
    else: # go left
      s_idx -= 1

    if s_idx == -1:
      episode.append(0)
      break
    elif s_idx == 5:
      episode.append(1)
      break
    else:
      episode.append(0)
      episode.append(state[s_idx])
  
  return episode

# initializing value functions
def init_val():
  value = [0.5, 0.5, 0.5, 0.5, 0.5]

# Monte-Carlo
def MC(iteration_num = 10):
  init_val()
  for iter in range(iteration_num):
    episode = generate_episode()
    for i in range(0, len(episode), 2):
      s = state_dict[episode[i]]
      ret = 0
      for j in range(i+1, len(episode), 2):
        ret += episode[j]
      value[s] = round(value[s] + alpha * (ret - value[s]), 2)

# Temporal-Difference
def TD_0(iteration_num = 10):
  init_val()
  for iter in range(iteration_num):
    episode = generate_episode()
    for i in range(0, len(episode), 2):
      s = state_dict[episode[i]]
      reward = episode[i+1]
      if i+2 < len(episode):
        v_next = state_dict[episode[i+2]]
        value[s] = round(value[s] + alpha * (reward + gamma * value[v_next] - value[s]), 2)
      else:
        value[s] = round(value[s] + alpha * (reward - value[s]), 2)
```

#### 3. Run & Result


```python
optimal_value = [round(1/6, 2), round(2/6, 2), round(3/6, 2), round(4/6, 2), round(5/6, 2)]
print(f"Optimal Value function : {optimal_value}")

N = 6 # 6 is enough

# MC Results
iter = 10
print("\n---- Monte-Carlo ----")
for i in range(N):
  start = time.time()
  MC(iteration_num=iter)
  print(f"Value function with MC : {value} , iter_nume : {iter}, ELAP : {time.time() - start:.5f}")
  iter *= 10

# TD Results
iter = 10
print("\n---- Tenporal-Difference ----")
for i in range(N):
  start = time.time()
  TD_0(iteration_num=iter)
  print(f"Value function with TD : {value} , iter_nume : {iter}, ELAP : {time.time() - start:.5f}")
  iter *= 10
```
